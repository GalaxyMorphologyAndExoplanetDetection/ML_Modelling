```python
import pandas as pd
from keras.models import Model, load_model
from keras.preprocessing.image import ImageDataGenerator
import sys
sys.path.insert(0, 'D:/OneDrive/Major Project/Code/Galaxy_Morphology/Data/RSA/utils')
import utils
from keras.models import load_model
```

```python
model = load_model('D:/OneDrive/Major Project/HybridModel_37Classes/params/00007-test-train/model-36-0.15.hdf5')
```

```python
test_datagenerator = ImageDataGenerator(rescale=1/255)

test_generator = test_datagenerator.flow_from_directory(
    directory="D:/OneDrive/Major Project/Code/Galaxy_Morphology/Data/RSA",
    classes=['validated_images'],
    class_mode=None,
    color_mode="rgb",
    batch_size=1,
    target_size=(299, 299),
    seed=123,
    shuffle=False)
```

```python
predictions = model.predict(
    test_generator,
    steps=test_generator.n/test_generator.batch_size,
    verbose=1)

print(predictions)

predictions.shape
```

```python
df = pd.DataFrame(predictions)
df

df.to_csv('D:/OneDrive/Major Project/Code/Galaxy_Morphology/Data/GalaxyZoo2/GZ2_37_vectors_predictions.csv')
```

```python
def displayPredictions(df):
    galaxyClasses = []
    for i in range(len(df)):  
        gClass = utils.classify(list(df.iloc[i, :]))
        galaxyClasses.append(gClass)
        print(i, " - ", gClass)
    return galaxyClasses
    
```

```python
galaxyType = displayPredictions(df)

rsa = pd.read_csv('D:/OneDrive/Major Project/Code/Galaxy_Morphology/Data/RSA/ProcessedRSA.csv')
rsa = rsa[rsa['classes'].notnull()]
rsa['Predictions'] = galaxyType
rsa
```

```python
print(rsa.loc[1246, 'Galaxy'])
print(rsa.loc[1246, 'classes'])
```

```python
pred = ['E0', 'E1', 'E2', 'E3', 'E4', 'E5', 'E7', 'SBx', 'Sx']
for i in range(1,len(rsa)+1):
    if rsa.loc[i, 'classes'] == 'E' and (rsa.loc[i, 'Predictions'] in pred ):
        rsa.loc[i, 'classes'] = rsa.loc[i, 'Predictions']
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
accuracy_score(list(rsa['classes']), list(rsa['Predictions']))

classification_report(list(rsa['classes']), list(rsa['Predictions']))

confusion_matrix(list(rsa['classes']), list(rsa['Predictions']))
```
